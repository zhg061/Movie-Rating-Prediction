```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn import linear_model
from scipy.linalg import svd
from sklearn.model_selection import KFold
```

```python
df = pd.read_csv('data/movies_ratings.csv')
```

```python
df = df.drop(columns=['Unnamed: 0', 'timestamp'])
df['rating_avg'] = df['rating_avg'].apply(lambda x: 2*x)
```

```python
df.columns[240:374]
```

```python
df = df.dropna()
clf = linear_model.Lasso(alpha=0.1)
X = df.drop(columns=['rating_avg', 'vote_count','vote_average'])
y = df.rating_avg
res = clf.fit(X, y)

coef = clf.coef_

features = X.columns[coef!=0]
print(features)
print(len(X.columns))
print(len(features)) 
```
