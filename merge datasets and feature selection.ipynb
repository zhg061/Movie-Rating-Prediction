```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
movie = pd.read_csv('data/movies_metadata_last12.csv')
rate = pd.read_csv('data/ratings_small.csv')
link = pd.read_csv('data/links_small.csv')
```

```python
movie = movie.drop(columns=['Unnamed: 0'])
movie.head()
```

```python
#join with rate's movieId with link's movieId
# join movie's imdb_id with link's imdbID
rate.columns, movie.columns, link.columns
```

```python
link = link.rename(columns={"imdbId": "imdb_id"})
link.columns
```

```python
rate.shape, movie.shape, link.shape
```

```python
rate_link = pd.merge(rate, link, how="inner", on="movieId")
rate_link.shape
```

```python
rate_link
```

```python
movie.dropna(subset = ["imdb_id"], inplace=True)
movie.shape
```

```python
movie.imdb_id = [int(x[-7:]) for x in movie.imdb_id]
```

```python
df = pd.merge(rate_link, movie, how="inner", on="imdb_id")
print(df.shape)
df.head()
```

```python
print(df.shape)
df = df.drop(columns=['userId', 'movieId', 'imdb_id', 'id', 'tmdbId'])
print(df.shape)
```

```python
# df.to_csv("data/movies_ratings.csv")
```

```python
df
```

```python
df = df.dropna()
clf = linear_model.Lasso(alpha=0.1)
X = df.drop(columns=['rating'])
y = df.rating
res = clf.fit(X, y)

coef = clf.coef_

features = X.columns[coef!=0]
print(features)
print(len(X.columns))
print(len(features)) 
```
