```python
# packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import linalg
from scipy.sparse.linalg import svds
import statsmodels.api as sm
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score
from sklearn import linear_model

import warnings
warnings.simplefilter('ignore', np.RankWarning)
```

```python
data = pd.read_csv("data/movies_ratings.csv", index_col="Unnamed: 0")
print(data.shape)
data = data.drop(columns=['revenue', 'vote_count', 'vote_average', 'Released', 'popularity'])
print(data.shape)
data = data.dropna()
data
```

## Standardize the Data
- Compute the standard deviation (s.d.) of each of the predictors and make a plot showing them. 

```python
# Compute the standard deviation (s.d.) of each of the predictors
std = []
predictors = data.columns[1:]
print(predictors)
for i in predictors:
    std.append(np.std(data[i]))    

# make a plot showing them
plt.figure(figsize=(18,5)) 
plt.scatter(predictors, std)
plt.xlabel("Predictors")
plt.ylabel("Standard Deviation")
plt.xticks(predictors[::20],  rotation='45')
plt.show()
```

```python
std
```

## standardize the predictors 
- by dividing each predictor by its standard deviation. 
- This ensures that all predictors have the same variance. 
- Using these values, define a new set of standardized predictors $A_{ij} =\frac{X_{ij}}{std(X_i)}$. 
- Check that you have done this correctly by verifying that the s.d. of each of the standardized predictors is 1.

```python
# define a new set of standardized predictors  𝐴
A = pd.DataFrame()
for i in range(len(predictors)):
    col = predictors[i]
    if std[i] == 0:
        A[col] = data[col]
    else:
        A[col] = data[col]/std[i]
A.head()   
```

```python
# Check that i have done this correctly
for i in predictors:
    std_i = np.std(A[i])
    if np.round(std_i) != 1:
        print("incorrect at: ", i)
```

## Perform singular value decomposition (SVD) 
- on the data matrix, A, by using scipy.linalg.svd. 
- SVD decomposes the data matrix, A, as a product of three matrices, $A=USV^T$.

```python
n = A.shape[0]
p = A.shape[1]

# Perform singular value decomposition (SVD) 
U, s, Vh = linalg.svd(A, full_matrices=False)
S = np.diag(s)

# What are the sizes (i.e. “shapes”) of the three matrices
print(U.shape,  S.shape, Vh.shape)

# How do these relate to the size of the original data matrix, A
np.allclose(A, np.dot(U, np.dot(S, Vh)))
```

##  Run PC regression 
- by fitting a linear model to predict y (“Behavior response”)using the top k=1 principal component to predict the behavioral outcome. 
    - (Hint: You could use statsmodels.regression.linear_model.OLS). 
- Plot the model’s prediction for the behavior as a function of time. 
- What is the MSE of this model fit?

```python
pc1 = np.dot(A.T, U[:, 0])
pc2 = np.dot(A.T, U[:, 1])
pc3 = np.dot(A.T, U[:, 2])

# Make a plot showing the first 3 columns of U.
plt.figure(figsize=(18,5)) 
plt.plot(predictors, pc1, label="PC1")
plt.plot(predictors, pc2, label="PC2")
plt.plot(predictors, pc3, label="PC3")
plt.xticks(predictors[::10],  rotation='45')
plt.legend()
plt.xlabel("Frequency Bands")
plt.ylabel("Coefficient")
plt.show()

# Which features contribute most strongly
pc1_freq = predictors[list(abs(pc1)).index(max(abs(pc1)))]
pc2_freq = predictors[list(abs(pc2)).index(max(abs(pc2)))]
pc1_freq, pc2_freq
```

## Run PC regression 
- by fitting a linear model to predict y (“Behavior response”)using the top k=[1, ..., 10] principal component to predict the behavioral outcome. 

```python
ks = np.arange(1, 378)
res = pd.DataFrame()
res['k'] = ks

Y = data['rating_avg']
mse = []
cv_score = []
V = Vh.T

for k in ks:    
    # Compute the SVD
    X = A @ V[:, :k]

    # Run PC regression by fitting a linear model to predict y
    ols = linear_model.LinearRegression()
    model = sm.OLS(Y,X)#.fit()
    y_pred = model.fit().predict(X)

    # the MSE of this model fit
    mse.append(mean_squared_error(Y, y_pred))
    # cross validation scores
    cv_5_results = cross_val_score(ols, 
                                   X, 
                                   Y, 
                                   cv=5, 
                                   scoring="neg_mean_squared_error")
    cv_score.append(np.mean(cv_5_results))
res['cv_score'] = cv_score
res['mse'] = mse
res
```

```python
res[res.cv_score == max(res.cv_score)]
```

```python
fig,ax = plt.subplots()

x = res.k
y1 = res.cv_score/np.std(res.cv_score)
y2 = res.mse/np.std(res.mse)

plt.plot(x, y1, marker='o', label="CV-score")
plt.plot(x, y2, marker='x', label="MSE")
plt.xticks(x[::43],  rotation='45')
plt.axvline(x = 44, color = 'g', linestyle = '-', label="Optimal K value")
plt.xlabel("K")
plt.ylabel("Score")
plt.title("K values in PCA vs CV Scores & Testing MSE")
plt.legend()
plt.show()
```
